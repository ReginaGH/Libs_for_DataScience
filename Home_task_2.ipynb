```python
a = np.arange(0, 12)
a
```

```python
At = A.transpose()
At, At.shape
```

`Ответ`:

```python
c = np.random.choice(np.arange(0, 16, 1), 16, replace=False)
c
```

### 12. 
Поменяйте его форму так, чтобы получилась квадратная матрица C. Получите матрицу D,
поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10.
Вычислите определитель, ранг и обратную матрицу D_inv для D

```python
C = c.reshape(4,4)
C
```

```python
B
```

```python
D = B + C * 10
D
```

```python
d = np.linalg.det(D)
r = np.linalg.matrix_rank(D)
print(f'Определитель: {d}, Ранг: {r}')
```

```python
# Определитель отличен от 0, значит можно посчитать обратную матрицу:
D_inv = np.linalg.inv(D)
D_inv
```
